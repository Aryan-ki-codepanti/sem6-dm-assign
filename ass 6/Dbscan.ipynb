```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# data
np.random.seed(30)
x = np.random.randint(1,60,20)
y =np.random.randint(1,60,20)
data = list(zip(x,y))
print(data)
plt.scatter(x,y)
```

```python
db = DBSCAN(eps=17, min_samples=5).fit(data)
```

```python
db.labels_
```

```python
db.core_sample_indices_
```

```python
plt.scatter(x,y,c=db.labels_)
```
