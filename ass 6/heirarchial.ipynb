```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# data
np.random.seed(30)
x = np.random.randint(1,60,20)
y =np.random.randint(1,60,20)
data = list(zip(x,y))
print(data)
plt.scatter(x,y)
```

```python
agg = AgglomerativeClustering(n_clusters=4,compute_full_tree=True).fit(data)
```

```python
agg.labels_
```

```python
agg.n_connected_components_
```

```python
agg.n_leaves_
```

```python
linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```
