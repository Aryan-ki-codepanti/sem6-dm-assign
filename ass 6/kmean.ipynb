```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
# data
np.random.seed(30)
x = np.random.randint(1,60,20)
y =np.random.randint(1,60,20)
data = list(zip(x,y))
print(data)
plt.scatter(x,y)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0, n_init="auto").fit(data)
```

```python
kmeans.labels_
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(x,y,c=kmeans.labels_)
plt.scatter(centers[:,0],centers[:,1], marker="*",s=100 , label="Centers")
plt.legend(loc="best")
```
